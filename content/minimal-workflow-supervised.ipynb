## Minimal Workflow: Supervised Learning


### 1. Load Dataset


```python
from sklearn.datasets import load_wine

ds = load_wine()

X = ds.data  # Features (wine chemical properties)
y = ds.target  # Labels (wine classes)

# Display Dataset description
print(ds.DESCR)
```

### 2. Split data


```python
from sklearn.model_selection import train_test_split

# 2. Split data into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=0.2, random_state=0
)

print("Original amount:", len(X))
print("Train amount:", len(X_train))
print("Test amount:", len(X_test))
```

### 3. Initialize model and train


```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=0)
model.fit(X_train, y_train)
```

### 4. Make predictions


```python
y_pred = model.predict(X_test)
```

### 5. Evaluate Model


```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Visualize the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix")
plt.show()
```
