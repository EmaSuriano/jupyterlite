## Minimal Workflow: Supervised Learning


### 1. Load Dataset


```python
from sklearn.datasets import load_wine

ds = load_wine()

X = ds.data  # Features (wine chemical properties)
y = ds.target  # Labels (wine classes)

# Display Dataset description
print(ds.DESCR)
```

### 2. Split data


```python
from sklearn.model_selection import train_test_split

# 2. Split data into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

### 3. Initialize model and train


```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=0)
model.fit(X_train, y_train)
```

### 4. Make predictions and evaluate


```python
from sklearn.metrics import accuracy_score


y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("Model accuracy:", accuracy)
```
